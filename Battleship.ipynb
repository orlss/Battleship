```python
class Battleship(object):
    def __init__(self):
        self.board=[]
        self.row=5
        self.column=5
        self.ship=[]
        self.shipspots=0
        self.shiphits=0
    def makeBoard(self):
        for i in range(0,self.row):
            self.board.append([])
            self.ship.append([])
            for j in range(0,self.column):
                self.board[i].append("O")
                self.ship[i].append("O")
    def printBoard(self):
        for i in range(0,self.row):
            print(self.board[i])
    def printShip(self):
        for i in range(0,self.row):
            print(self.ship[i])
    def changeBoard(self,x,y):
        self.board[y][x]="X"
    def makeShip(self):  
        print("make a ship:")
        x=input("input x:")
        while(x.isdigit()==False):
            print("enter an integer value for x")
            x=input("input x:")
        x=int(x)
        
        y=input("input y:")
        while(y.isdigit()==False):
            print("enter an integer value for y")
            y=input("input y:")
        y=int(y)
        
        size=input("input size:")
        while(size.isdigit()==False):
            print("enter an integer value for size")
            size=input("input size:")
        size=int(size)
        
        orientation=input("input orientation:")
        while(orientation.isdigit()==False):
            print("enter an integer value for orientation")
            orientation=input("input orientation:")
        orientation=int(orientation)

        if (orientation==0):
            if (x+size>5):
                print("ship is out of bounds")
                self.makeShip()
            else: 
                ship_exists=False
                for k in range(size):
                    if(self.ship[y][x+k]=="S"):
                        ship_exists=True
                if(ship_exists==False):
                    for i in range(size):
                        self.ship[y][x+i]="S"
                    self.shipspots=self.shipspots+size
                else:
                    print("There is already a ship here, try again!")
                    self.makeShip()
        else:
            if (y+size>5):
                print("ship is out of bounds")
                self.makeShip()
            else: 
                ship_exists=False
                for k in range(size):
                    if(self.ship[y+k][x]=="S"):
                        ship_exists=True
                if(ship_exists==False):
                    for i in range(size):
                        self.ship[y+i][x]="S"
                    self.shipspots=self.shipspots+size
                else:
                    print("There is already a ship here, try again!")
                    self.makeShip()
#        if (self.ship=="S"):  We need to work on this part
#            print("a ship already exists at this coordinate. please input new coordinates")
#            self.makeShip()
    def hitShip(self):
        print("hit a ship:")
        x=input("input x:")
        while(x.isdigit()==False):
            print("enter an integer value for x")
            x=input("input x:")
        x=int(x)
        
        y=input("input y:")
        while(y.isdigit()==False):
            print("enter an integer value for y")
            y=input("input y:")
        y=int(y)
        
        if (self.board[y][x]!="O"):
            print("you've already guessed this spot","\n")
        else:
            if (self.ship[y][x]=="S"):
                print("hit ship at",x,".",y,"\n")
                self.board[y][x]="X"
                self.shiphits=self.shiphits+1
            else: 
                print("miss ship at",x,".",y,"\n")
                self.board[y][x]="M"
        self.printBoard()
        if (self.shipspots==self.shiphits):
            print("you win! all ships have been hit","\n","GAME OVER")
# Need to switch x and y coordinates for hitship
    def playGame(self):
#        to begin game, type in (battleship=Battleship())
        self.makeBoard()
        self.printBoard()
        
        a=int(input("enter how many ships you would like to make:"))
        for i in range(a):
            self.makeShip()
        
        while(self.shipspots!=self.shiphits):
            self.hitShip()
```
